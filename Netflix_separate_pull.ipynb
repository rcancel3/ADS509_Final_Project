```python
import pandas as pd

# for scraping app info from App Store
from itunes_app_scraper.scraper import AppStoreScraper

# for scraping app reviews from App Store
from app_store_scraper import AppStore

# for pretty printing data structures
from pprint import pprint

# for keeping track of timing
import datetime as dt
from tzlocal import get_localzone

# for building in wait times
import random
import time
```

```python
## Instantiate AppStore class for desired app
my_app = AppStore(
  country='us',        # required, 2-letter code
  app_name='netflix', # required, found in app's url
  app_id=363590051    # technically not required, found in app's url
) 
    
## Use review method to scrape reviews from App Store
my_app.review(
    how_many=1000, # if not provided, defaults to scraping all
    after=dt.datetime(2022, 1, 1),
    sleep=random.randint(20,25))        # optional, int, seconds between each call
```

```python
# Save the Netflix reviews
netflix = my_app.reviews

# Convert list of dicts to Pandas DataFrame and write to csv
netflix_df = pd.DataFrame(netflix)
netflix_df['app_name'] = 'netflix'
netflix_df['app_id'] = 363590051
netflix_df.to_csv('C:/Users/rober/OneDrive - University of San Diego/Documents/GitHub/ADS509_Final_Project/Data/' + 'netflix' + '.csv', index=False)
    
```
